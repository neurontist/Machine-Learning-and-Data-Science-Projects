```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1,2,3]
y = [4,5,6]

# Resize your Graph
plt.figure(figsize=(5,4),dpi=100)


# plt.plot(x,y, label='first graph', color='red', linewidth=0.9, marker = 'o'
#          ,markersize=2,markeredgecolor='blue',linestyle='--')

# Another way to write the same
# USE SHORTHAND NOTATION
# fmt = '[color][marker][line]'
plt.plot(x,y,'r>--',label = 'My First Graph')


### Line Number Two
x2 = np.arange(0,4,0.5)
plt.plot(x2[:6],x2[:6]**2,'b',label='X^2')
plt.plot(x2[5:],x2[5:]**2,'b--')

plt.title(label="My First Graph",fontdict = {'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X axis', fontdict = {'fontname':'Arial'})
plt.ylabel('Y axis', fontdict = {'fontname':'Arial'})

# plt.xticks([0,1])
plt.yticks([0,2,4,6,8,10])

plt.legend()

# plt.savefig('my first grpah.png',dpi=300)  #saving the graph
plt.show()
```

### Bar Graph


```python
labels = ['A','B','C']
values = [1,8,4]

plt.figure(figsize=(5,3),dpi=300)

bars = plt.bar(labels,values, label=['India','UK','USA'])


# bars[0].set_hatch('/')
# bars[1].set_hatch('*')
# bars[2].set_hatch('o')

patterns = ['/','*','o']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.legend()
plt.show()
```

### Fifa Data

```python
fifa = pd.read_csv('sample_data/fifa_data.csv')
fifa.head(5)
```

### Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins , color="lightblue")

plt.xticks(bins)
plt.xlabel('Number of Players')
plt.ylabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

### Pie Charts

```python
labels = ['left','right']
num_foot = [(fifa['Preferred Foot']=='Left').sum(),(fifa['Preferred Foot']=='Right').sum()]
num_foot

colors = ['pink','lightblue']

plt.pie(num_foot, labels=labels, autopct='%.2f%%', colors=colors)
plt.title('Foot Preference of FIFA Players')
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

fifa['Weight'] = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >=150) & (fifa.Weight <175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >=175)&(fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >=200)].count()[0]

weights = [light, light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-150','150-175','175-200','Over200']
explode = (0,0.1,0,0,0)
plt.title('Weight Distribution of FIFA Players (in lbs)')
plt.pie(weights,labels=labels, autopct='%.2f%%', pctdistance=0.7, explode=explode )
plt.show()
```

```python
plt.style.use('default')

barcelona = fifa[fifa['Club'].str.contains('Barcelona', na=False)]['Overall']
madrid = fifa[fifa['Club'].str.contains('Madrid', na=False)]['Overall']

labels = ['Barcelona','Madrid']

plt.boxplot([barcelona,madrid], labels=labels)

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.xlabel('Teams')
plt.show()
```

## Subplot

```python
x = np.arange(100)

fig , axs = plt.subplots(2,2)

axs[0,0].plot(x,x**2)
axs[0,1].plot(x,x**3)
axs[1,0].plot(x,x**4)
axs[1,1].plot(x,x**5)

axs[0,0].set_title('X^2')
axs[0,1].set_title('X^3')
axs[1,0].set_title('X^4')
axs[1,1].set_title('X^5')

axs[0,0].set_xlabel('X')
axs[0,1].set_xlabel('X')
axs[1,0].set_xlabel('X')
axs[1,1].set_xlabel('X')

axs[0,0].set_ylabel('Y')
axs[0,1].set_ylabel('Y')
axs[1,0].set_ylabel('Y')
axs[1,1].set_ylabel('Y')

plt.show()
```

```python

ax = plt.axes(projection='3d')

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter3D(x,y,z,c=z,cmap='Greens')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
%matplotlib inline

# Create a 3D axes
fig = plt.figure()  # Create a figure object
ax = fig.add_subplot(111, projection='3d')  # Add 3D subplot

# Generate data
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='Spectral')

# Add labels and title
ax.set_title('3D Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# ax.view_init(elev=90, azim=0)

# Show the plot
plt.show()
```
